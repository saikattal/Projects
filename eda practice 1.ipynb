```python
cl_sales2=cl_sales1.set_index('Order ID').drop('Unnamed: 0',axis=1)
```

```python
cl_sales2
```

```python
cl_sales2.groupby('Order Date')['Sales'].agg(['mean','count'])
```

```python
cl_sales2.pivot_table('Sales',index='Product',columns='City')
```

```python
cl_sales2['Zip Code']=cl_sales2['Purchase Address'].str.split(' ').str[-1]
```

```python
cl_sales2.head()
```

```python
cl_sales2['Dist Code']=cl_sales2['Purchase Address'].str.split(' ').str[-2]
```

```python
cl_sales2['Month'].replace({1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'Jun',7:'Jul',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'},inplace=True)
```

```python
#help(pd.to_datetime(cl_sales2['Order Date']).dt)
```

```python
cl_sales2['Year']=pd.to_datetime(cl_sales2['Order Date']).dt.year
```

```python
cl_sales2['Month']=pd.to_datetime(cl_sales2['Order Date']).dt.month
```

```python
cl_sales2['Month Name']=pd.to_datetime(cl_sales2['Order Date']).dt.month_name()
```

```python
cl_sales2['Day of Week']=pd.to_datetime(cl_sales2['Order Date']).dt.dayofweek
```

```python
cl_sales2['Day of Week Name']=pd.to_datetime(cl_sales2['Order Date']).dt.day_name()
```

```python
cl_sales2['Order Date']=pd.to_datetime(cl_sales2['Order Date'])
```

```python
monthly_sales=cl_sales2.set_index('Order Date')['Sales'].resample('2W').mean()
```

```python
sns.lineplot(data=monthly_sales,x=monthly_sales.index,y=monthly_sales.values)
```
